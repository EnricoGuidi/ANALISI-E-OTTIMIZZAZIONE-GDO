```python
import pandas as pd

# === 1. CARICAMENTO FILE EXCEL ===
# Sostituire il percorso con quello corretto dove si trova il file .xlsx sul proprio sistema
file_path = "/Users/enricoguidi/Downloads/Copia di analisi_localizzazione_liguria.xlsx"

# Caricamento del file Excel nel dataframe 'df_caricato'
df_caricato = pd.read_excel(file_path, sheet_name="Sheet1")

display(df_caricato)

# === 2. CREAZIONE DEL DATAFRAME CON I KPI ===

# Selezione delle colonne rilevanti dal file Excel
df_kpi = df_caricato[[
    'Località',
    'Popolazione totale',
    'Reddito familiare mediano',
    'Densità di popolazione',
    'Indice di spesa dei consumatori',
    'Numero di concorrenti diretti',
    'Distanza Concorrenza (km)',
    'Costo Apertura (Mln €)',
    'Indice di accessibilità al trasporto pubblico',
    'Tasso di crescita del mercato locale'
]].copy()

# === 3. CALCOLO KPI DERIVATO ===

# Calcolo dell'indice sintetico di attrattività economica
df_kpi["Indice sintetico di attrattività economica"] = (
    df_kpi["Popolazione totale"] * df_kpi["Reddito familiare mediano"]
) / df_kpi["Costo Apertura (Mln €)"]

# (Opzionale) Visualizzazione delle prime righe per verifica
display(df_kpi.head())

# (Opzionale) Salvataggio su file Excel o CSV per analisi successive
# df_kpi.to_excel("output_kpi_localizzazione.xlsx", index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler

# Selezione delle colonne da normalizzare
colonne_da_normalizzare = [
    'Popolazione totale',
    'Reddito familiare mediano',
    'Densità di popolazione',
    'Indice di spesa dei consumatori',
    'Numero di concorrenti diretti',
    'Distanza Concorrenza (km)',
    'Costo Apertura (Mln €)',
    'Indice di accessibilità al trasporto pubblico',
    'Tasso di crescita del mercato locale',
    'Indice sintetico di attrattività economica'
]

# Pesi
pesi = {
    'Popolazione totale': 0.20,
    'Reddito familiare mediano': 0.15,
    'Densità di popolazione': 0.10,
    'Indice di spesa dei consumatori': 0.10,
    'Numero di concorrenti diretti': 0.05,
    'Distanza Concorrenza (km)': 0.10,
    'Costo Apertura (Mln €)': 0.15,
    'Indice di accessibilità al trasporto pubblico': 0.05,
    'Tasso di crescita del mercato locale': 0.05,
    'Indice sintetico di attrattività economica': 0.05
}

# Copia del DataFrame per normalizzazione
df_norm = df_kpi.copy()
scaler = MinMaxScaler()

# Normalizzazione
df_norm[colonne_da_normalizzare] = scaler.fit_transform(df_kpi[colonne_da_normalizzare])

# Inversione dei criteri con impatto negativo
df_norm['Numero di concorrenti diretti'] = 1 - df_norm['Numero di concorrenti diretti']
df_norm['Costo Apertura (Mln €)'] = 1 - df_norm['Costo Apertura (Mln €)']

# Calcolo del punteggio totale ponderato
df_norm['Punteggio Totale'] = sum(df_norm[col] * peso for col, peso in pesi.items())

# Ordinamento per punteggio
df_result = df_norm[['Località', 'Punteggio Totale']].sort_values(by='Punteggio Totale', ascending=False)

# Visualizzazione del grafico con barre verticali
plt.figure(figsize=(10, 6))
barplot = sns.barplot(data=df_result, x='Località', y='Punteggio Totale', palette='viridis')
plt.title('Punteggio Totale per Località (Normalizzato e Ponderato)')
plt.ylabel('Punteggio Totale')
plt.xlabel('Località')

# Aggiunta dei valori sopra le barre
for container in barplot.containers:
    barplot.bar_label(container, fmt='%.3f', padding=3)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from math import pi

# 1. HEATMAP DELLE CORRELAZIONI
plt.figure(figsize=(12, 8))
corr = df_kpi.drop(columns=["Località"]).corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Heatmap di Correlazione tra KPI")
plt.tight_layout()
plt.show()

# 2. RADAR CHART - TOP 3 LOCALITÀ
from sklearn.preprocessing import MinMaxScaler

# Normalizzazione KPI
kpi_cols = [
    'Popolazione totale',
    'Reddito familiare mediano',
    'Densità di popolazione',
    'Indice di spesa dei consumatori',
    'Numero di concorrenti diretti',
    'Distanza Concorrenza (km)',
    'Costo Apertura (Mln €)',
    'Indice di accessibilità al trasporto pubblico',
    'Tasso di crescita del mercato locale',
    'Indice sintetico di attrattività economica'
]

df_norm = df_kpi.copy()
scaler = MinMaxScaler()
df_norm[kpi_cols] = scaler.fit_transform(df_kpi[kpi_cols])
df_norm['Numero di concorrenti diretti'] = 1 - df_norm['Numero di concorrenti diretti']
df_norm['Costo Apertura (Mln €)'] = 1 - df_norm['Costo Apertura (Mln €)']


df_norm["Punteggio Totale"] = sum(df_norm[col] * peso for col, peso in pesi.items())

# Selezione Top 3 località
top3 = df_norm.nlargest(3, "Punteggio Totale").set_index("Località")

# Radar chart
# Modifica delle etichette con testo a capo per migliorare leggibilità nel radar chart
labels_wrapped = [
    'Popolazione\ntotale',
    'Reddito\nfamiliare\nmediano',
    'Densità\ndi\npopolazione',
    'Indice di\nspesa dei\nconsumatori',
    'Numero di\nconcorrenti\ndiretti',
    'Distanza\nConcorrenza\n(km)',
    'Costo Apertura\n(Mln €)',
    'Indice di\naccessibilità al\ntrasporto pubblico',
    'Tasso di\ncrescita del\nmercato locale',
    'Indice sintetico\ndi attrattività\neconomica'
]

angles = [n / float(len(labels_wrapped)) * 2 * pi for n in range(len(labels_wrapped))]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(10, 8), subplot_kw=dict(polar=True))
for name, row in top3.iterrows():
    values = row[kpi_cols].tolist()
    values += values[:1]
    ax.plot(angles, values, label=name)
    ax.fill(angles, values, alpha=0.1)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels_wrapped, fontsize=9)
ax.set_title("Radar Chart – Top 3 Località (10 KPI)", size=14)
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
plt.tight_layout()
plt.show()
```

## SENSITIVITY ANALYSIS – andamento VARIAZIONE PESO delle variabili

```python
# Estensione del modello a Multi-Obiettivo
# L'obiettivo è esplorare scenari multipli e verificare la variazione del punteggio in funzione di diverse combinazioni di pesi

# Preparazione dati
df_multi = df_kpi.copy()

# Uso delle stesse colonne già definite per la normalizzazione
# 'colonne_da_normalizzare' è già disponibile dal primo blocco
df_multi[colonne_da_normalizzare] = scaler.fit_transform(df_multi[colonne_da_normalizzare])

# Inversione criteri negativi (già definiti nel primo blocco)
df_multi['Numero di concorrenti diretti'] = 1 - df_multi['Numero di concorrenti diretti']
df_multi['Costo Apertura (Mln €)'] = 1 - df_multi['Costo Apertura (Mln €)']

# Generazione scenari multi-obiettivo
scenario_results = []
weight_steps = [0.3, 0.25, 0.2, 0.15, 0.1, 0.05]

for var_prioritaria in colonne_da_normalizzare:
    for wp in weight_steps:
        remaining_weight = 1.0 - wp
        other_vars = [v for v in colonne_da_normalizzare if v != var_prioritaria]
        n = len(other_vars)
        uniform_weight = remaining_weight / n
        new_weights = {v: uniform_weight for v in other_vars}
        new_weights[var_prioritaria] = wp

        # Calcolo punteggio
        df_temp = df_multi.copy()
        df_temp["Punteggio Multi-Obiettivo"] = sum(df_temp[col] * new_weights[col] for col in colonne_da_normalizzare)
        top_locality = df_temp.loc[df_temp["Punteggio Multi-Obiettivo"].idxmax(), "Località"]

        scenario_results.append({
            "Variabile Prioritaria": var_prioritaria,
            "Peso Assegnato": wp,
            "Top Località": top_locality,
            "Punteggio Massimo": df_temp["Punteggio Multi-Obiettivo"].max()
        })

# DataFrame dei risultati
df_scenari = pd.DataFrame(scenario_results)

# Visualizzazione del comportamento delle località ottimali per ogni scenario
plt.figure(figsize=(14, 6))
sns.lineplot(data=df_scenari, x="Peso Assegnato", y="Top Località", hue="Variabile Prioritaria", marker="o")
plt.title("Scenari Ottimali Multi-Obiettivo – Impatto della Priorità delle Variabili")
plt.xlabel("Peso Assegnato alla Variabile Prioritaria")
plt.ylabel("Località con Punteggio Massimo")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()


```

```python
# ================================
# 🧠 Analisi Avanzata – Frequenza di Scelta della Località Ottimale
# ================================

# 'df_scenari' contiene già per ogni combinazione (Variabile Prioritaria, Peso Assegnato) la località vincente

# Conteggio della frequenza con cui ogni località è risultata la migliore
df_localita_frequenza = df_scenari['Top Località'].value_counts().reset_index()
df_localita_frequenza.columns = ['Località', 'Frequenza di Scelta']

# Visualizzazione con valori sulle barre
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=df_localita_frequenza, x='Frequenza di Scelta', y='Località', palette='crest')
plt.title("Frequenza con cui ciascuna Località è risultata la Migliore")
plt.xlabel("Numero di Scenari Ottimali")
plt.ylabel("Località")

# Aggiunta dei valori numerici sopra le barre
for i, (valore, localita) in enumerate(zip(df_localita_frequenza['Frequenza di Scelta'], df_localita_frequenza['Località'])):
    ax.text(valore + 0.5, i, str(valore), va='center')

plt.tight_layout()
plt.show()


```

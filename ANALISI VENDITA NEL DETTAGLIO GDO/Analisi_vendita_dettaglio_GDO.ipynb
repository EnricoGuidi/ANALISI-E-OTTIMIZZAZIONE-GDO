## <div class="header2">5.1 | Gestione dei Dati "Sporchi" üßπ</div>
<div class="explain-box">
    Nella sezione precedente √® stato gi√† menzionato che <b>sono stati rilevati alcuni valori incoerenti nella colonna "Item_Fat_Content"</b>. In questa sezione, <mark>quei valori incoerenti verranno trasformati in un unico valore</mark>, con lo stesso significato/interpretazione.
</div>

```python
# Verifica dei valori mancanti prima della pulizia
missing_values = df.isnull().sum()
print("Valori mancanti prima della pulizia:")
print(missing_values)

# Imputazione dei valori mancanti:
# - Per le colonne numeriche usiamo la mediana.
# - Per le colonne categoriche usiamo il valore pi√π frequente.
for col in df.columns:
    if df[col].dtype in ['float64', 'int64']:
        df[col].fillna(df[col].median(), inplace=True)
    else:
        df[col].fillna(df[col].mode()[0], inplace=True)

print("\nValori mancanti dopo la pulizia:")
print(df.isnull().sum())
```

```python
# Tentativo di conversione delle colonne in formato numerico ove possibile
for col in df.columns:
    try:
        df[col] = pd.to_numeric(df[col])
    except Exception:
        # Se la conversione non √® possibile, la colonna viene lasciata invariata
        pass

print("Tipi di dati dopo eventuale conversione:")
print(df.dtypes)
```

### 5.2.1 | Outlet_Size

Prima di procedere con il processo di imputazione, **verr√† effettuata un'analisi preliminare** per assicurarsi che i valori da imputare siano corretti, evitare risultati distorti e garantire un dataset di alta qualit√†. 

In base all‚Äôanalisi svolta nella sezione precedente, **gli outlet di dimensioni medie risultano essere i pi√π numerosi** nella distribuzione. 

> Tuttavia, un‚Äôimputazione con il valore "medio" renderebbe il dataset distorto, poich√© la dimensione media finirebbe per dominare sulle altre dimensioni degli outlet.

Per questo motivo, il primo passo sar√† quello di **raggruppare i tipi di outlet e le loro dimensioni, per poi calcolare il numero totale di outlet disponibili nel dataset**.


```python
import numpy as np

# --- Calcolo del numero di prodotti unici per Tipo e Dimensione del punto vendita ---
df.groupby(['Outlet_Type', 'Outlet_Size'], dropna=False)['Item_Identifier'] \
  .nunique() \
  .reset_index() \
  .sort_values(by=['Outlet_Size'], ascending=False) \
  .style.background_gradient(cmap='GnBu')

# --- Imputazione: sostituisco NaN in Outlet_Size con "Small" ---
df['Outlet_Size'] = df['Outlet_Size'].replace(np.nan, 'Small')

# --- Visualizzazione post-imputazione ---
df.groupby(['Outlet_Type', 'Outlet_Size'], dropna=False)['Item_Identifier'] \
  .nunique() \
  .reset_index() \
  .sort_values(by=['Outlet_Size'], ascending=False) \
  .style.background_gradient(cmap='Blues')
```

<div class="explain-box">
    Dalla tabella sopra, <mark>la dimensione "piccola" del punto vendita √® associata al negozio di alimentari e, in larga parte, al supermercato di tipo 1</mark>. In questo caso, <mark>al negozio di alimentari verr√† assegnato il valore "piccolo"</mark> (utilizzando il metodo della sostituzione con l‚Äôultimo valore), mentre <mark>ai supermercati di tipo 1 verr√† imputata la modalit√† (valore pi√π frequente) della loro dimensione, che risulta essere "piccola"</mark>.
</div>


```python
# --- Performing Imputation ---
df['Outlet_Size'] = df['Outlet_Size'].replace(np.nan, 'Small')
# --- DataFrame dopo l‚Äôimputazione ---
df_grouped = df.groupby(['Outlet_Type', 'Outlet_Size'], dropna=False)['Item_Identifier'] \
               .nunique() \
               .reset_index() \
               .sort_values(by=['Outlet_Size'], ascending=False)

# Visualizzazione con stile solo se in ambiente notebook
try:
    from IPython.display import display
    display(df_grouped.style.background_gradient(cmap='Blues'))
except ImportError:
    # Se non in Jupyter, stampa normale
    print(df_grouped)
```

## 5.3 | Gestione degli Outlier üîß

> Come menzionato nella sezione precedente, **"Item_Visibility"** e **"Item_Outlet_Sales"** presentano outlier nella loro distribuzione.  
> In questa sezione ci si concentrer√† sulla **trasformazione di queste colonne** per eliminare gli outlier utilizzando **trasformazioni logaritmiche e radice quadrata**.  
> Dopo aver effettuato la trasformazione, **i risultati delle tecniche verranno confrontati inizialmente** prima di essere applicati al dataset reale.


```python
# --- Creating Transformation DataFrame ---
df_transform = df[['Item_Visibility', 'Item_Outlet_Sales']].copy()

# --- Log Transformation ---
df_transform['Item_Visibility_log'] = np.log(df_transform['Item_Visibility'])
df_transform['Item_Outlet_Sales_log'] = np.log(df_transform['Item_Outlet_Sales'])

# --- Using Square Root Transformation
df_transform['Item_Visibility_sqrt'] = np.sqrt(df_transform['Item_Visibility'])
df_transform['Item_Outlet_Sales_sqrt'] = np.sqrt(df_transform['Item_Outlet_Sales'])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Definizione colori e font ---
color_line = 'black'  # Colore per gli elementi di linea
color_grid = 'gray'  # Colore per la griglia
colors = ['blue', 'green', 'red', 'purple', 'orange']  # Palette di colori
font_main = 'Arial'  # Font principale
font_alt = 'Arial'  # Font alternativo

# --- DataFrame (assicurati che df_transform sia gi√† definito) ---
df_transform_skew = pd.DataFrame(df_transform.skew(axis=0, skipna=True)).T
df_transform_kurt = pd.DataFrame(df_transform.kurt(axis=0, skipna=True)).T

# --- Variabili ---
ax_num_iv, ax_num_os = 0, 0
tick_params = dict(length=3, width=1, color=color_line)
suptitle = dict(x=0.5, y=0.94, fontsize=20, weight='heavy', ha='center', va='bottom', fontname=font_main)
title = dict(x=0.5, y=0.93, fontsize=15, weight='normal', ha='center', va='bottom', fontname=font_alt)
highlight_textprops = [{'weight': 'bold', 'color': colors[1]}, {'weight': 'bold'}, {'weight': 'bold'}]

# --- Funzione di supporto per il grafico ---
def distribution_plot(ax, col_name):
    skew_value = df_transform_skew.iloc[0][col_name]
    kurt_value = df_transform_kurt.iloc[0][col_name]

    ax.xaxis.set_label_position('top')
    ax.set_title(f"\n{col_name}", fontsize=13, fontweight='bold')
    ax.set_xlabel(f'Skewness Value: {skew_value:.4f}\nKurtosis Value: {kurt_value:.4f}',
                  fontsize=10, fontname=font_alt)
    ax.set_ylabel('')
    ax.grid(axis='y', alpha=0, zorder=2)
    ax.grid(axis='x', which='major', alpha=0.3, color=color_grid, linestyle='dotted', zorder=1)
    
    # Nascondere le spine del grafico tranne quella inferiore
    for spine in ax.spines.values():
        spine.set_color('None')
    ax.spines['bottom'].set_visible(True)
    ax.spines['bottom'].set_color(color_line)

    ax.tick_params(bottom=True, **tick_params)

# --- Creazione dei grafici ---
fig, ax = plt.subplots(3, 2, figsize=(19, 21))
for col, axes in zip(list(df_transform.columns), ax.flatten()):
    if col.startswith("Item_Visibility"):
        sns.histplot(data=df_transform, x=col, kde=True, ax=axes, color=colors[3])
        distribution_plot(axes, col)
        ax_num_iv += 1
    else:
        sns.histplot(data=df_transform, x=col, kde=True, ax=axes, color=colors[-2])
        distribution_plot(axes, col)
        ax_num_os += 1

# --- Titoli e descrizioni ---
plt.suptitle('Transformation Comparison', **suptitle)
plt.title('<Distribution Comparison after Transformation> on <"Item_Visibility"> and <"Item_Outlet_Sales">',
          fontdict=title)

plt.gcf().text(0.78, 0.07, 'kaggle.com/caesarmario', style='italic', fontsize=12)
fig.tight_layout()
plt.show()

```

<div class="explain-box">
    Sulla base degli istogrammi, dei valori di skewness e kurtosis mostrati sopra, <mark>la trasformazione tramite radice quadrata √® migliore rispetto alla trasformazione logaritmica</mark>, poich√© <b>le distribuzioni degli istogrammi con trasformazione a radice quadrata sono pi√π vicine alla distribuzione normale</b>, anche se sono ancora presenti outlier in entrambe le colonne. Inoltre, il <b>valore di skewness ha mostrato che la trasformazione a radice quadrata pu√≤ ridurre efficacemente la skewness</b>. ## <div class="header2">8.2 | Raggruppamento in Classi (Binning)</div>

```python
import pandas as pd


# Controllo della presenza della colonna 'Age_Outlet'
if "Age_Outlet" not in df.columns:
    raise ValueError("La colonna 'Age_Outlet' non √® presente nel dataset.")

# --- Creazione della colonna 'Outlet_Category' mediante binning ---
bins = [1, 10, 20, 100]  # Definizione delle classi
outlet_labels = ['New Outlet', 'Moderate Outlet', 'Old Outlet']  # Categorie corrispondenti
df['Outlet_Category'] = pd.cut(df['Age_Outlet'], bins=bins, labels=outlet_labels)

# --- Visualizzazione del DataFrame completo ---
print("\n‚úÖ Dataset aggiornato con la colonna 'Outlet_Category':\n")
print(df)  # Mostra il DataFrame intero nel terminale di VSCode
```

    In questa sezione verr√† applicata la <mark>codifica delle etichette (label encoding)</mark> alla colonna <b>"Item_Fat_Content"</b>.
</div>


```python

# Controllo della presenza della colonna 'Item_Fat_Content'
if "Item_Fat_Content" not in df.columns:
    raise ValueError("La colonna 'Item_Fat_Content' non √® presente nel dataset.")

# --- Applicazione del Label Encoding su 'Item_Fat_Content' ---
df['Item_Fat_Content_Encoding'] = df['Item_Fat_Content'].replace({'Low Fat': '0', 'Regular': '1'})

# --- Visualizzazione del DataFrame completo ---
print("\n‚úÖ Dataset aggiornato con la colonna 'Item_Fat_Content_Encoding':\n")
df  # Mostra il DataFrame intero nel terminale di VSCode
```

## <div class="header2">8.4 | One-Hot Encoding</div>
<div class="explain-box">
    Verr√† applicato il <mark>One-Hot Encoding</mark> alle colonne <b>"Item_Category", "Outlet_Category", "Outlet_Size" e "Outlet_Location_Type"</b>.
</div>


```python
# Definizione delle colonne da trasformare con One-Hot Encoding
onehot_cols = ['Item_Category', 'Outlet_Category', 'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type']

# Controllo delle colonne presenti nel dataset
missing_cols = [col for col in onehot_cols if col not in df.columns]
if missing_cols:
    raise ValueError(f"Le seguenti colonne necessarie per il One-Hot Encoding sono mancanti nel dataset: {missing_cols}")

# --- Esecuzione del One-Hot Encoding ---
df = pd.get_dummies(data=df, columns=onehot_cols, drop_first=True)

# Conversione esplicita delle colonne codificate in int64
binary_columns = [col for col in df.columns if df[col].dtype == 'bool']
df[binary_columns] = df[binary_columns].astype(int)

# Verifica del tipo di dati dopo la conversione
print(df.dtypes)
```

## <div class="header2">8.5 | Eliminazione delle Variabili Non Necessarie</div>
<div class="explain-box">
    Le <mark>variabili non necessarie verranno eliminate</mark>. Inoltre, questa sezione <mark>mostrer√† il DataFrame finale dopo il processo di feature engineering</mark>.
</div>


```python

# Controllo delle colonne prima dell'eliminazione
cols_to_drop = ['Item_Fat_Content', 'Outlet_Establishment_Year']
missing_cols = [col for col in cols_to_drop if col not in df.columns]
if missing_cols:
    raise ValueError(f"Le seguenti colonne non sono presenti nel dataset: {missing_cols}")

# --- Eliminazione delle colonne non necessarie ---
df = df.drop(columns=cols_to_drop)

# --- Visualizzazione del DataFrame aggiornato ---
print("\n‚úÖ Dataset aggiornato dopo la rimozione delle colonne:\n")
print(df)  # Mostra il DataFrame intero nel terminale di VSCode
```
